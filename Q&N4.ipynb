# Matrix multiplication.

```python
import numpy as np
```

```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

print(np.dot(a,b))
```

# Searching & Sorting

```python
arr = np.array([10,20,30,20])

print(np.where(arr==20))
```

# Count non-zero values.

```python
arr = np.array([1,0,2,0,3])

print(np.count_nonzero(arr))
```
